# Example 2: Load dataset, subset by cluster mapping, save subsets

This notebook demonstrates `subset_adata_by_cluster_mapping` with `Cnn -> subset` mapping.

```python
from pathlib import Path
import scomnom as om

# Edit these paths for your environment
in_path = Path("/path/to/input_dataset.zarr")
out_dir = Path("/path/to/subsets")
out_dir.mkdir(parents=True, exist_ok=True)
```

```python
adata = om.load_dataset(in_path)
adata
```

```python
subset_mapping = {
    "C00": "Parenchymal",
    "C01": "Endothelial",
    "C02": "Parenchymal",
    "C03": "Stromal",
    "C04": "Immune",
    "C05": "Epithelial",
}

subsets, summary = om.adata_ops.subset_adata_by_cluster_mapping(
    adata,
    subset_mapping=subset_mapping,
)
summary
```

```python
for name, sub in subsets.items():
    out_path = out_dir / f"subset_{name}.zarr"
    om.save_dataset(sub, out_path, fmt="zarr", archive=True)
    print(f"Saved {name}: {out_path}")
```
